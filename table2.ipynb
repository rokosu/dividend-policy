```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Set style
sns.set_theme(style="whitegrid")
#plt.style.use('seaborn-whitegrid')
sns.set_palette("husl")

# Create synthetic beta coefficients based on research findings
data = {
    'Determinant': ['Regulatory Compliance', 'Forex Scarcity', 'Profitability (ROE)', 'Leverage'],
    'Banking': [0.28, -0.73, 0.34, 0.18],
    'Manufacturing': [0.10, -0.65, 0.30, -0.27],
    'Telecom': [0.15, -0.60, 0.32, 0.05]
}

df = pd.DataFrame(data)

# Melt data for visualization
melted_df = df.melt(id_vars='Determinant',
                    var_name='Sector',
                    value_name='Beta Coefficient')

# Create figure
plt.figure(figsize=(14, 8))

# Plot with error bars (simulated confidence intervals)
bar = sns.barplot(x='Determinant',
                 y='Beta Coefficient',
                 hue='Sector',
                 data=melted_df,
                 ci="sd",  # Simulated error bars
                 errwidth=1.5,
                 capsize=0.1)

# Add annotations
for p in bar.patches:
    height = p.get_height()
    if height != 0:
        bar.annotate(f"{height:.2f}",
                    (p.get_x() + p.get_width() / 2., height),
                    ha='center',
                    va='center',
                    xytext=(0, 9),
                    textcoords='offset points',
                    fontsize=10)

# Customize plot
plt.title("Figure 4: Determinants of Payout Ratios by Sector (Î² Coefficients)",
         fontsize=14, pad=20)
plt.xlabel("Determinants", fontsize=12, labelpad=15)
plt.ylabel("Standardized Beta Coefficients", fontsize=12, labelpad=15)
plt.axhline(0, color='black', linewidth=1)
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)
plt.legend(title='Sector', title_fontsize=12, fontsize=11)

# Add theoretical explanation
plt.text(0.5, -0.85,
        "Key Insights:\n"
        "1. Regulatory pressures strongest in banking sector (CAM Act compliance)\n"
        "2. Forex scarcity shows universal negative impact across sectors\n"
        "3. Leverage effects diverge: positive in banking (regulation), negative in manufacturing",
        ha='center', va='center', fontsize=12,
        bbox=dict(facecolor='white', alpha=0.8))

plt.tight_layout()
plt.savefig('sector_determinants.png', dpi=300, bbox_inches='tight')
plt.show()
```
